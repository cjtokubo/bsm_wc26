```python
subset = my_data[['TeamName','TotalDistance','DistanceHighSpeedRunning','DistanceHighSpeedSprinting','DistanceJogging','DistanceLowSpeedSprinting','DistanceWalking']]
subset['%Slow'] = subset['DistanceWalking']/subset['TotalDistance']
subset['%Fast'] = (subset['DistanceHighSpeedRunning'] + subset['DistanceHighSpeedSprinting'])/subset['TotalDistance']
```
